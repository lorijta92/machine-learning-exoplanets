```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://www.breadahead.com/live-bake"
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
# Search for recipe title container
results = soup.find_all("div", class_="user_content")
```

```python
results[1]
```

```python
# Search for recipe title
titles = results[1].find_all('h2')
```

```python
print(titles)
```

```python
for title in titles:
    print(title.text)
```
